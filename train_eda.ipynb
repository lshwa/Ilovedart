```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 사기 여부 비율 출력
fraud_counts = df['fraud'].value_counts()
fraud_percent = df['fraud'].value_counts(normalize=True) * 100

print("사기 여부 (fraud) 분포:")
print(fraud_counts)
print("\n사기 여부 비율 (%):")
print(fraud_percent.round(2))

# 시각화
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='fraud', palette='Set2')
plt.title('Fraud Label Distribution')
plt.xlabel('Fraud (0:정상, 1:사기)')
plt.ylabel('Count')
plt.xticks([0, 1], ['정상(0)', '사기(1)'])
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 결측치 많아 제외하거나 -1로 대체
df['initial_transfer_amount'].fillna(-1, inplace=True)
df['prev_address_months_count'].fillna(-1, inplace=True)

# 적당한 결측치: 평균으로 대체
df['bank_months_count'].fillna(df['bank_months_count'].mean(), inplace=True)
df['current_address_months_count'].fillna(df['current_address_months_count'].mean(), inplace=True)
df['session_length_in_minutes'].fillna(df['session_length_in_minutes'].mean(), inplace=True)
df['device_distinct_emails_8w'].fillna(df['device_distinct_emails_8w'].mean(), inplace=True)
```

```python
df.isnull().sum().sum()  # 결과가 0이면 모든 결측치 처리 완료
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 수치형 변수만 골라서 fraud와의 상관계수 보기
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.drop('fraud')
correlations = df[numeric_cols].corrwith(df['fraud']).sort_values(key=abs, ascending=False)

# 상위 10개 시각화
plt.figure(figsize=(8, 6))
sns.barplot(x=correlations.head(10), y=correlations.head(10).index, palette='coolwarm')
plt.title('Top 10 Correlated Features with Fraud')
plt.xlabel('Correlation with fraud')
plt.show()
```

```python
for col in ['payment_type', 'source', 'employment_status', 'housing_status', 'email_is_free']:
    print(f"\n▶ {col} 별 사기율")
    print(df.groupby(col)['fraud'].mean().sort_values(ascending=False))
```

```python
# income vs fraud 시각화 (수치형 변수)
sns.boxplot(x='fraud', y='income', data=df)
plt.title('Income Distribution by Fraud')
plt.show()
```
