### 결측치 처리

```python
# 중앙값 대체
df_train['bank_months_count'].fillna(df_train['bank_months_count'].median(), inplace=True)
df_train['current_address_months_count'].fillna(df_train['current_address_months_count'].median(), inplace=True)
df_train['session_length_in_minutes'].fillna(df_train['session_length_in_minutes'].median(), inplace=True)

# 평균 대체
df_train['device_distinct_emails_8w'].fillna(df_train['device_distinct_emails_8w'].mean(), inplace=True)

# 0으로 대체
df_train['prev_address_months_count'].fillna(0, inplace=True)
df_train['initial_transfer_amount'].fillna(0, inplace=True)
```

### 인코딩

```python
categorical_cols = df_train.select_dtypes(include=['object', 'bool']).columns.tolist()
```

변수명	자료형	고유값 수	예시 값	추천 인코딩 방식	이유

device_os	object	5	linux, windows 등	One-Hot	OS 간 서열 없음

employment_status	object	~6	employed, retired 등	One-Hot	명목형, 비교 불가능

housing_status	object	~6	own, rent 등	One-Hot	명목형, 의미 없음

payment_type	object	~6	card, cash 등	One-Hot	명목형

source	object	2	INTERNET, TELEAPP	Binary Encoding 또는 One-Hot	2개라면 binary 가능

email_is_free	bool	2	True / False	0/1 매핑 (그대로 사용)	이진 변수

foreign_request	bool	2	True / False	0/1 매핑 (그대로 사용)	이진 변수

has_other_cards	bool	2	True / False	0/1 매핑 (그대로 사용)	이진 변수

keep_alive_session	bool	2	True / False	0/1 매핑 (그대로 사용)	이진 변수

phone_home_valid	bool	2	True / False	0/1 매핑 (그대로 사용)	이진 변수

phone_mobile_valid	bool	2	True / False	0/1 매핑 (그대로 사용)	이진 변수



```python
# 인코딩 대상 구분
one_hot_cols = ['device_os', 'employment_status', 'housing_status', 'payment_type', 'source']
binary_cols = ['email_is_free', 'foreign_request', 'has_other_cards',
               'keep_alive_session', 'phone_home_valid', 'phone_mobile_valid']

# 복사본 생성
df_encoded = df_train.copy()

# Boolean 변수 0/1 변환
df_encoded[binary_cols] = df_encoded[binary_cols].astype(int)

# One-Hot 인코딩 적용
df_encoded = pd.get_dummies(df_encoded, columns=one_hot_cols, drop_first=False)

# 결과 미리보기
print(df_encoded.head())

```
