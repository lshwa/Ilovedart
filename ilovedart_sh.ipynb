```python
import seaborn as sns

numeric_cols = file_3.select_dtypes(include=['int64', 'float64']).columns.drop('fraud')
file_3[numeric_cols] = file_3[numeric_cols].fillna(file_3[numeric_cols].median())
corr_with_fraud = file_3[numeric_cols].corrwith(file_3['fraud']).abs().sort_values(ascending=False)
top_features = corr_with_fraud.head(10).index.tolist()

plt.figure(figsize=(10, 6))
sns.barplot(
    x=corr_with_fraud[top_features],
    y=top_features,
    palette='viridis',
    legend=False  # FutureWarning 방지
)
plt.title("상위 fraud와 연관된 10개의 컬럼")
plt.xlabel("fraud와의 절대 상관계수")
plt.ylabel("컬럼")
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# top_features 대상 
selected_features = top_features + ['fraud']  

corr_matrix = file_3[selected_features].corr()

plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # 상삼각형 마스크

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    square=True,
    linewidths=.5,
    cbar_kws={"shrink": .8}
)

plt.title("fraud 및 상위 연관 피처 간 상관관계 히트맵 (부분)")
plt.tight_layout()
plt.show()
```

```python
# fraud 여부별 credit_risk_score 
file_3.groupby('fraud')['credit_risk_score'].agg(['mean', 'median', 'std', 'min', 'max', 'count'])
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='fraud', y='credit_risk_score', data=file_3)
plt.title('Boxplot of credit_risk_score by fraud')

# KDE Plot 그래프 credit_risk_score와 사기와의 연관성
plt.subplot(1, 2, 2)
sns.kdeplot(data=file_3[file_3['fraud'] == 0]['credit_risk_score'], label='Not Fraud', fill=True)
sns.kdeplot(data=file_3[file_3['fraud'] == 1]['credit_risk_score'], label='Fraud', fill=True)
plt.title('KDE of credit_risk_score by fraud')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 사기인 경우 credit_risk_score vs fraud 분석 결과 
# fraud가 1일 때가 더 평균 / 중앙값이 훨씬 높아보임. 
# 사기 가능성이 더 높아 보인다. 
```

```python
# fraud 여부별 credit_risk_score 
file_3.groupby('fraud')['requested_credit_limit'].agg(['mean', 'median', 'std', 'min', 'max', 'count'])
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='fraud', y='requested_credit_limit', data=file_3)
plt.title('Boxplot of credit_risk_score by fraud')

# KDE Plot 그래프 requested_credit_limit와 사기와의 연관성
plt.subplot(1, 2, 2)
sns.kdeplot(data=file_3[file_3['fraud'] == 0]['requested_credit_limit'], label='Not Fraud', fill=True)
sns.kdeplot(data=file_3[file_3['fraud'] == 1]['requested_credit_limit'], label='Fraud', fill=True)
plt.title('KDE of credit_risk_score by fraud')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# credit_risk_score vs fraud 분석 결과 
# fraud가 1일 때가 더 평균 / 중앙값이 훨씬 높아보임. 
# 사기 사용자들이 평균적으로 더 높은 한도를 요청한다. 
```

```python
# IQR 기준 계산
q1 = file_3['credit_risk_score'].quantile(0.25)
q3 = file_3['credit_risk_score'].quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr

# 조건 적용: 이상치이면서 정상인 경우
nonfraud_high_outliers = file_3[(file_3['credit_risk_score'] > upper_bound) & (file_3['fraud'] == 0)]

# 통계 출력
print(f"정상인데 이상치인 경우 개수: {len(nonfraud_high_outliers)}")
print(nonfraud_high_outliers['credit_risk_score'].describe())
```

```python
# 전체 정상 유저
nonfraud = file_3[file_3['fraud'] == 0]

# credit_risk_score 이상치인 정상 유저
nonfraud_outliers = nonfraud[nonfraud['credit_risk_score'] > upper_bound]

# 예: income 비교
nonfraud[['income']].describe()
nonfraud_outliers[['income']].describe()

# 또는 requested_credit_limit 등 반복
```

```python
nonfraud_df = file_3[file_3['fraud'] == 0]
numeric_cols = nonfraud_df.select_dtypes(include=['int64', 'float64']).columns.drop('fraud')
nonfraud_df[numeric_cols] = nonfraud_df[numeric_cols].fillna(nonfraud_df[numeric_cols].median())
corr_matrix = nonfraud_df[numeric_cols].corr()
mean_corr = corr_matrix.abs().mean().sort_values(ascending=False)
top_features = mean_corr.head(10).index.tolist()

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(
    x=mean_corr[top_features],
    y=top_features,
    palette='viridis',
    legend=False
)
plt.title("정상 사용자 내부에서 상호 상관계수 높은 10개 컬럼")
plt.xlabel("다른 변수들과의 평균 절대 상관계수")
plt.ylabel("컬럼")
plt.tight_layout()
plt.show()
```

```python
selected_features = top_features + ['fraud']

corr_matrix = file_3[selected_features].corr()

plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # 상삼각 마스크

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": .8}
)

plt.title("fraud 및 상위 연관 피처 간 상관관계 히트맵 (부분)")
plt.tight_layout()
plt.show()
```

```python
# fraud 의 전체에서의 분포 확인 
fraud_counts = file_3['fraud'].value_counts()
fraud_ratio = file_3['fraud'].value_counts(normalize=True) * 100

print("Fraud Class 분포:")
print(fraud_counts)
print("\n Fraud Class 비율 (%):")
print(fraud_ratio.round(4))
```
